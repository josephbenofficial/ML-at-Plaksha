```python
# Import the requried libraries
import numpy as np
import keras
from keras.datasets import cifar 10
from keras.models import Sequential
from keras.layers import Dense,Conv2D, Flatten, MaxPooling2D
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
```

```python
# loading the CIFAR-10 datasets
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
for i in range(9):
    plt.subplot(330 + 1 + i)
    plt.imshow(X_train[i])
plt.show()
```

```python
# Let's do one-hot encoding as the outputs are represented by integer values
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Normalize the training set
X_train_norm = X_train.astype('float32') / 255.0
X_test_norm = X_test.astype('float32') / 255.0
```

```python
# Making the CNN Model
model = Sequential()
model.add(Conv2D(64, kernel_size = 3, activation = "relu", input_shape = (32,32,3)))
model.add(Conv2D(128,kernel_size = 3, activation = "relu"))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(256,kernel_size = 3, activation = "relu"))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(512,kernel_size = 3, activation = "relu"))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(10,activation = 'softmax'))
```

```python
# Stating the optimizer, loss function and metric to be used.
model.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=1)
for (train,test), i in zip(cv.split(X_train_norm, y_train), range(5)):
     model.fit(X_train_norm[train], y_train[train], epochs = 2)
```

```python
model.evaluate(X_test,y_test)
```
